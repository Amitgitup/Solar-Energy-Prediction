```python
import numpy as np
import pandas as pd 
```

```python
df = pd.read_csv('Solar_Energy_Prediction.csv')
df.head()
```

```python
## Z-Score
from scipy import stats

z_scores = np.abs(stats.zscore(df))
df_no_outliers = df[(z_scores <3).all(axis=1)]
```

```python
df_no_outliers
```

```python
df_no_outliers
```

```python
## IQR

Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)

IQR = Q3 - Q1

df_iqr = df[~((df < (Q1 - 1.5*IQR)) | (df > (Q3 + 1.5*IQR))).any(axis=1)]
```
