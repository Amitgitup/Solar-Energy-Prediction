```python
## IQR

Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)

IQR = Q3 - Q1

df_iqr = df[~((df < (Q1 - 1.5*IQR)) | (df > (Q3 + 1.5*IQR))).any(axis=1)]
df_iqr
```

```python
## Splitting the features into Independent and Dependent Features
X = df_iqr.iloc[:,:-1]
y = df_iqr.iloc[:,-1]

## Splitting the data into training data and testing data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=1)

## Loading 'scaler.pkl' file 
with open('scaler.pkl', 'rb') as file:
    scaler = pickle.load(file)

## Performing Standardization
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## ANN Implementation

```python
import tensorflow 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard
import datetime
```

```python
(X_train.shape[1], )
```

```python
## Create a Sequential Network
model = Sequential([
    Dense(128, activation='relu', input_shape=(X_train.shape[1],)),       ## HL1
    Dense(128, activation='relu'),
    Dense(128, activation='relu'),
    # Dense(64, activation='relu'),                                        ## HL2
    Dense(1, activation='linear')                                        ## Output Layer
])
```

```python
model.summary()
```

```python
## Compile the model
model.compile(optimizer='adam', loss='huber', metrics=['mae'])
```

```python
## Setup the TensorBoard
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard

log_dir = 'logs/fit/' + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = TensorBoard(log_dir = log_dir, histogram_freq=1)
```

```python
## Setup the EarlyStopping
early_stopping_callback = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
```

```python
## Model Training 
history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=100,
    validation_split=0.2,
    batch_size=64,
    callbacks= [tensorboard_callback, early_stopping_callback]
)
```

```python
test_loss, test_mae = model.evaluate(X_test, y_test)
print(f'Test MAE: {test_mae}')
```

### Hyperparameter Tuning ANN

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from scikeras.wrappers import KerasRegressor
```

```python
def create_model(neurons=32, layers=1):
    model1 = Sequential()
    model1.add(Dense(neurons, activation='relu', input_shape=(X_train.shape[1],)))

    for _ in range(layers-1):
        model1.add(Dense(neurons, layers-1))

    model1.add(Dense(1, activation='linear'))
    model1.compile(optimzer='adam', loss='mean_absolute_error', metrics=['mae'])    

    return model1
```

```python
## Create a KerasRegressor
model1 = KerasRegressor(neurons=32, layers=1, model=create_model, verbose=1)
```

```python
## Define the grid search parameters 
param_grid = {
    'neurons': [32, 48, 64, 128 ],
    'layers': [1, 2 ,3],
    'eopchs': [50, 100]
}
```

```python
from sklearn.model_selection import GridSearchCV

grid = GridSearchCV(estimator=model1, param_grid=param_grid, cv=3, n_jobs=-1, verbose=1)
grid.fit(X_train, y_train)
```
